Loading Dependencies


```python
import pandas as pd
```

Reading Dataset

```python
df = pd.read_csv("Disease_symptom.csv")
df.head()
```

```python
df.info()
```

**Data Cleaning**

```python
df.isnull().sum()
```

```python
drop_df = df.dropna()
df=drop_df
drop_df.isnull().sum()
```

**Find Duplicates**

```python
duplicate_mask = df.duplicated(keep=False)
duplicate_rows = df[duplicate_mask]
```

```python
print(duplicate_rows)
```

**Removing Duplicates**

```python
df_unique = df.drop_duplicates()
print("DataFrame after removing duplicates:")
print(df_unique)
```

```python
df=df_unique
print(df)
```

Binary Encoding

```python
df['Gender'].unique()
df['Gender'] = df['Gender'].replace(['Male', 'Female'], [1, 0])

df['Fever'].unique()
df['Fever'] = df['Fever'].replace(['Yes', 'No'], [1, 0])

df['Cough'].unique()
df['Cough'] = df['Cough'].replace(['Yes', 'No'], [1, 0])

df['Fatigue'].unique()
df['Fatigue'] = df['Fatigue'].replace(['Yes', 'No'], [1, 0])

df['Difficulty Breathing'].unique()
df['Difficulty Breathing'] = df['Difficulty Breathing'].replace(['Yes', 'No'], [1, 0])

df['Outcome Variable'].unique()
df['Outcome Variable'] = df['Outcome Variable'].replace(['Positive', 'Negative'], [1, 0])
```

One-Hot Categorical Encoding


```python
df['Blood Pressure'].unique()
df = pd.get_dummies(df, columns=['Blood Pressure'], prefix='BP', dtype='int64')

df['Cholesterol Level'].unique()
df = pd.get_dummies(df, columns=['Cholesterol Level'], prefix='cholestrol', dtype='int64')
print(df)
```

```python
df.head()
```

Re-Ordering Columns

```python
desired_order = ['Disease','Fever', 'Cough', 'Fatigue', 'Difficulty Breathing','Age','Gender','BP_High','BP_Low','BP_Normal','cholestrol_High','cholestrol_Low','cholestrol_Normal',
                 'Outcome Variable']
df = df[desired_order]
```

CO-relation


```python
df.corr()
```

Cleaned And Processed File

```python
df.to_csv("Processed_Disease_symptom.csv", index=False)
```
